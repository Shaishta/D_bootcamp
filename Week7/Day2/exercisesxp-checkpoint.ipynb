```python
#2.
import pandas as pd
import matplotlib.pyplot as plt
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
plt.figure(figsize= (10,5))
titanic_df.Fare.plot(kind="hist", color = 'g')
#Fare distribution
plt.xlabel('Fare')
plt.show()
```

```python
first_c = titanic_df[titanic_df.Pclass == 1].Survived.value_counts(normalize=True)
second_c = titanic_df[titanic_df.Pclass == 2].Survived.value_counts(normalize=True)
third_c = titanic_df[titanic_df.Pclass == 3].Survived.value_counts(normalize=True)


fig, ax = plt.subplots(1,3, figsize = (13,6))

ax[0].pie(first_c, labels = first_c.index.map({0: "Not Survived", 1: "Survived"}), autopct = '%1.1f%%',)
ax[0].set_title('First_class')
ax[1].pie(second_c, labels = second_c.index.map({0: "Not Survived", 1: "Survived"}), autopct = '%1.1f%%',)
ax[1].set_title('Second_class')

ax[2].pie(third_c, labels = third_c.index.map({0: "Not Survived", 1: "Survived"}), autopct = '%1.1f%%',)
ax[2].set_title('Third_class')

plt.show()
```

```python
# age distribution
titanic_df.Age.plot(kind='hist')
plt.xlabel('Age')
plt.title('Age distribution')
plt.show()
```

```python
# age ditribution of survived
titanic_df.groupby('Survived')['Age'].plot(kind='kde')
plt.legend()
plt.title('Age by density by survived')
plt.show()
```

```python
# survived by gender
male_data = titanic_df[titanic_df.Sex == 'male'].Survived.value_counts(normalize=True)
female_data = titanic_df[titanic_df.Sex == 'female'].Survived.value_counts(normalize=True)
fig, ax = plt.subplots(1,2,figsize=(13,6))
ax[0].pie(male_data, labels = male_data.index.map({0: "Did Not Survive", 1: "Survived"}), autopct = '%1.1f%%',)
ax[0].set_title('Male Passengers')
ax[1].pie(female_data, labels = female_data.index.map({0: "Did Not Survive", 1: "Survived"}), autopct = '%1.1f%%',)
ax[1].set_title('Female Passengers')
```

```python
# survived by class num
first_class = titanic_df[titanic_df.Pclass == 1].Survived.value_counts(normalize=True)
second_class = titanic_df[titanic_df.Pclass == 2].Survived.value_counts(normalize=True)
third_class = titanic_df[titanic_df.Pclass == 3].Survived.value_counts(normalize=True)
fig, ax = plt.subplots(1,3,figsize=(13,6))

ax[0].pie(first_class, labels = first_class.index.map({0: "Did Not Survive", 1: "Survived"}), autopct = '%1.1f%%',)
ax[0].set_title('First Class')

ax[1].pie(second_class, labels = second_class.index.map({0: "Did Not Survive", 1: "Survived"}), autopct = '%1.1f%%',)
ax[1].set_title('Second Class')

ax[2].pie(third_class, labels = third_class.index.map({0: "Did Not Survive", 1: "Survived"}), autopct = '%1.1f%%',)
ax[2].set_title('Third Class')
```

```python
# survived by siblings on board, parents on board, fare
continuous_features = ['SibSp', 'Parch', 'Fare']
target = 'Survived'
fig, ax = plt.subplots(3,1,figsize=(8, 8))
for i, cont_v in enumerate(continuous_features):
    titanic_df.groupby(target)[cont_v].plot(kind='kde', ax=ax[i])
    ax[i].set_xlabel(cont_v)
    ax[i].legend()

fig.tight_layout()
plt.show()
```

```python
#3
titanic_df['Age'].fillna(value = titanic_df.Age.mean(), inplace = True)
titanic_df.Cabin.fillna(value='Unknown', inplace = True)
pd.get_dummies(titanic_df, columns=['Sex'])
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
df_scaled = titanic_df.copy()
col_names = ['Age','Fare']
features = df_scaled[col_names]
scaler = MinMaxScaler()
df_scaled[col_names] = scaler.fit_transform(features.values)
df_scaled
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
fig,ax=plt.subplots(1,2,figsize=(13, 5))
titanic_df['Age'].plot(kind = 'kde', ax=ax[0]);
ax[0].set_title('Age before Log Transform')
transformer = FunctionTransformer(np.log1p, validate=True)
pd.DataFrame(transformer.transform(titanic_df[['Age']])).plot(kind = 'kde', ax=ax[1]);
ax[1].set_title('Age after Log Transform')
plt.show()
```
